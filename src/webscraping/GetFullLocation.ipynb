```python
import pandas as pd
import requests
import json
import os
```

```python
df  = pd.read_csv('../datasets/vivareal.csv')
df
```

```python
def getPositionApiKey():
  with open('../../positionstack.json') as jsonFile:
    data = json.load(jsonFile)
    return data['API_KEY']
```

```python
def getLocation(queryParams):
  url = 'http://api.positionstack.com/v1/forward'
  params = {
    'access_key': getPositionApiKey(),
    'query': queryParams
  }
  response = requests.get(url, params=params)
  return response.json()
```

```python
df['latitude'] = None
df['longitude'] = None
df
```

```python
for i in range(len(df['location'])):
  locationResponse = getLocation(df['location'][i])
  
  if(len(locationResponse['data']) > 0):
    df['latitude'][i] = locationResponse['data'][0]['latitude']
    df['longitude'][i] = locationResponse['data'][0]['longitude']

    print(str(i) + ' - successfully add new location property: latitude:' + str(df['latitude'][i]) + ' longitude:' + str(df['longitude'][i]))
  else:
    print(str(i) + ' - no location found')
    
df
```

```python

import os
df.to_csv('../datasets/vivareal.csv', index = False, header=True)
```
