```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
wiki = 'https://www.cciimoveismg.com.br/comprar/mg/santa-rita-do-sapucai/casa/valor-min_0/area-min_0/ordem-valor/resultado-crescente/quantidade-92/'
page = urllib.request.urlopen(wiki)
soup = BeautifulSoup(page, 'html.parser')
```

```python
allHouses = soup.find('div', attrs={'class': 'todos_imoveis'})
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
```

```python
for row in allHouses.findAll('div', attrs={'class': ['resultado resultado_lista resultado_', 'resultado resultado_lista resultado_last' ]}): 
    A.append(row.find('h3', attrs={'class': 'tipo'}).text)
    B.append(row.find('h4', attrs={'class': 'localizacao'}).text)
    G.append(row.find('div', attrs={'class': 'valor'}).contents[1].text)
    
    details = row.find('div', attrs={'class': 'detalhes'})
    
    hasBedrooms = details.findChildren('i', attrs={'class': 'icon bed'})
    hasBathrooms = details.findChildren('i', attrs={'class': 'icon bath'})
    hasGarage = details.findChildren('i', attrs={'class': 'icon car'})
    
#     Se existir dados para todos os campos
    if(len(hasBedrooms) == 1 and len(hasBathrooms) == 1 and len(hasGarage) == 1):
        C.append(details.contents[0].text)
        D.append(details.contents[1].text)
        E.append(details.contents[2].text)
        F.append(details.contents[3].text)
#  Se existir dados para quartos, áreas e garagem, mas não existir para banheiro       
    elif(len(hasBedrooms) == 1 and len(hasBathrooms) == 0 and len(hasGarage) == 1):
        C.append(details.contents[0].text)
        D.append(0)
        E.append(details.contents[1].text)
        F.append(details.contents[2].text)
#  Se existir dados para área e quartos, mas não existir para banheiro e garagem      
    elif(len(hasBedrooms) == 1 and len(hasBathrooms) == 0 and len(hasGarage) == 0):
        C.append(details.contents[0].text)
        D.append(0)
        E.append(0)
        F.append(details.contents[1].text)
#  Se existir dados para quarto, área e banheiro, mas não existir para garagem       
    elif(len(hasBedrooms) == 1 and len(hasBathrooms) == 1 and len(hasGarage) == 0):
        C.append(details.contents[0].text)
        D.append(details.contents[1].text)
        E.append(0)
        F.append(details.contents[2].text)
#  Se existir dados para área, mas não existir para quarto, banheiro e garagem   
    elif(len(hasBedrooms) == 0 and len(hasBathrooms) == 0 and len(hasGarage) == 0):
        C.append(0)
        D.append(0)
        E.append(0)
        F.append(details.contents[0].text)
#  Se existir dados para área e banheiro, mas não existir para quarto e garagem   
    elif(len(hasBedrooms) == 0 and len(hasBathrooms) == 1 and len(hasGarage) == 0):
        C.append(0)
        D.append(details.contents[0].text)
        E.append(0)
        F.append(details.contents[1].text)
#  Se existir dados para área, banheiro e garagem, mas não existir para quarto    
    elif(len(hasBedrooms) == 0 and len(hasBathrooms) == 1 and len(hasGarage) == 1):
        C.append(0)
        D.append(details.contents[0].text)
        E.append(details.contents[1].text)
        F.append(details.contents[2].text)
#  Se existir dados para área e garagem, mas não existir para quarto e banheiro    
    elif(len(hasBedrooms) == 0 and len(hasBathrooms) == 0 and len(hasGarage) == 1):
        C.append(0)
        D.append(0)
        E.append(details.contents[0].text)
        F.append(details.contents[1].text)
```

```python
df = pd.DataFrame()
```

```python
df['Tipo']=A
df['Localidade']=B
df['Quarto']=C
df['Banheiro']=D
df['Garagem']=E
df['Área']=F
df['Preço']=G
df['Finalidade']='Venda'
```
